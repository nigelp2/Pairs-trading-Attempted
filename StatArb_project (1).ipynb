```python
import numpy as np
import pandas as pd
import yfinance as yf 
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('russell-2000-index-12-17-2025.csv')
tickers = df['Symbol'].dropna()
tickers = tickers.tolist()
test_tickers = tickers
test_tickers
```

```python
data = yf.download(test_tickers,  start='2024-12-17', end='2025-12-16', auto_adjust=True)['Close']
```

```python
returns = data.pct_change()
returns
```

```python
correlation_matrix = returns.corr()
print(correlation_matrix)
```

```python
for i in range(len(correlation_matrix)):
    for j in range(i+1, len(correlation_matrix)):
        corr_value = correlation_matrix.iloc[i, j]
        if corr_value > 0.85:
            df_pairs = (f"{correlation_matrix.columns[i]} - {correlation_matrix.columns[j]}: {corr_value:.2f}")
            print(df_pairs)
```

```python
pairs = []

# Loop through the upper triangle of the correlation matrix
for i in range(correlation_matrix.shape[0]):
    for j in range(i+1, correlation_matrix.shape[1]):
        corr_value = correlation_matrix.iloc[i, j]
        
        if corr_value > 0.85:
            pairs.append({
                "Stock1": correlation_matrix.columns[i],
                "Stock2": correlation_matrix.columns[j], 
                "Correlation": corr_value
            })

# Convert to DataFrame after the loop
df_pairs = pd.DataFrame(pairs)

print(df_pairs.head())
```

```python
selected_stocks = pd.unique(df_pairs[['Stock1', 'Stock2']].values.ravel())
selected_stocks
```

```python
z_scores = (close_subset - close_subset.mean()) / close_subset.std()
z_scores
```

```python
zscore_pairs_list = []

for _, row in df_pairs.iterrows():
    stock1 = row['Stock1']
    stock2 = row['Stock2']
    corr = row['Correlation']

    # Make sure both stocks exist in z_scores
    if stock1 in z_scores.columns and stock2 in z_scores.columns:
        df_temp = pd.DataFrame({
            "Date": z_scores.index,
            "Stock1": stock1,
            "Stock2": stock2,
            "Correlation": corr,
            "Zscore_Stock1": z_scores[stock1].values,
            "Zscore_Stock2": z_scores[stock2].values,
        })
        df_temp["Spread"] = df_temp["Zscore_Stock1"] - df_temp["Zscore_Stock2"]
        zscore_pairs_list.append(df_temp)

# Concatenate all pairs into one DataFrame
zscore_pairs_df = pd.concat(zscore_pairs_list, ignore_index=True)

print(zscore_pairs_df.head())
```

```python
ENTRY = 2.5
zscore_pairs_df['position'] = 0
zscore_pairs_df['Entry'] = 0
zscore_pairs_df['Exit'] = 0
```

```python
for t in range(1, len(zscore_pairs_df)):
    z = zscore_pairs_df.loc[zscore_pairs_df.index[t], 'Spread']
    prev_pos = zscore_pairs_df.loc[zscore_pairs_df.index[t-1], 'position']

    if z > ENTRY and prev_pos == 0:
        zscore_pairs_df.loc[zscore_pairs_df.index[t], 'position'] = -1
        zscore_pairs_df.loc[zscore_pairs_df.index[t], 'Entry'] = -1

    elif z < -ENTRY and prev_pos == 0:
        zscore_pairs_df.loc[zscore_pairs_df.index[t], 'position'] = 1
        zscore_pairs_df.loc[zscore_pairs_df.index[t], 'Entry'] = 1

    else:
        zscore_pairs_df.loc[zscore_pairs_df.index[t], 'position'] = prev_pos
        
```

```python
EXIT = 0.2
STOP = 3.0
if prev_pos != 0 and abs(z) < EXIT:
    zscore_pairs_df.loc[zscore_pairs_df.index[t], 'position'] = 0
    zscore_pairs_df.loc[zscore_pairs_df.index[t], 'Exit'] = 1
if prev_pos != 0 and abs(z) > STOP:
    zscore_pairs_df.loc[zscore_pairs_df.index[t], 'position'] = 0
    zscore_pairs_df.loc[zscore_pairs_df.index[t], 'position'] = 1
```

```python
stock1_list = zscore_pairs_df['Stock1'].values
stock2_list = zscore_pairs_df['Stock2'].values
z1_list = zscore_pairs_df['Zscore_Stock1'].values
z2_list = zscore_pairs_df['Zscore_Stock2'].values
corr_list = zscore_pairs_df['Correlation'].values

all_pairs = []

for i in range(len(stock1_list)):
    s1 = stock1_list[i]
    s2 = stock2_list[i]
    
    # Only include if returns exist
    if s1 in returns.columns and s2 in returns.columns:
        temp_df = pd.DataFrame({
            'Date': returns.index,
            'Stock1': s1,
            'Stock2': s2,
            'Returns_Stock1': returns[s1].values,
            'Returns_Stock2': returns[s2].values,
            'Zscore_Stock1': z1_list[i],
            'Zscore_Stock2': z2_list[i],
            'Spread': z1_list[i] - z2_list[i],
            'Weight_A': 1,
            'Weight_B': -1,  # You can calculate beta later if needed
            'Position': 0,
            'Position_A': 0,
            'Position_B': 0,
            'Entry': 0,
            'Exit': 0,
            'Stop': 0,
            'Correlation': corr_list[i]
        })
        all_pairs.append(temp_df)

# Concatenate only once
full_pair_df = pd.concat(all_pairs, ignore_index=True)
```

```python
all_pairs = []

for _, row in zscore_pairs_df.iterrows():
    stock1 = row['Stock1']
    stock2 = row['Stock2']

    if stock1 in returns.columns and stock2 in returns.columns:
        df_pair = pd.DataFrame({
            'Date': returns.index,
            'Stock1': stock1,
            'Stock2': stock2,
            'Returns_Stock1': returns[stock1].values,
            'Returns_Stock2': returns[stock2].values,
            'Zscore_Stock1': row['Zscore_Stock1'],
            'Zscore_Stock2': row['Zscore_Stock2'],
        })

        df_pair['Spread'] = df_pair['Zscore_Stock1'] - df_pair['Zscore_Stock2']

        # Calculate beta safely
        if np.var(df_pair['Zscore_Stock2'].values) > 0 and len(df_pair) > 1:
            beta = np.cov(df_pair['Zscore_Stock1'].values, df_pair['Zscore_Stock2'].values)[0, 1] \
                   / np.var(df_pair['Zscore_Stock2'].values)
        else:
            beta = 1.0  # default if variance = 0

        df_pair['Weight_A'] = 1
        df_pair['Weight_B'] = -beta

        df_pair['Position'] = 0
        df_pair['Position_A'] = 0
        df_pair['Position_B'] = 0
        df_pair['Entry'] = 0
        df_pair['Exit'] = 0
        df_pair['Stop'] = 0
        df_pair['Correlation'] = row['Correlation']

        all_pairs.append(df_pair)

full_pair_df = pd.concat(all_pairs, ignore_index=True).dropna()

print(full_pair_df.head())
```

```python
ENTRY = 2.5
EXIT = 0.2
STOP = 3.0

full_pair_df.loc[full_pair_df['Spread'] > ENTRY, 'Position'] = -1
full_pair_df.loc[full_pair_df['Spread'] < -ENTRY, 'Position'] = 1

# Exit
full_pair_df.loc[full_pair_df['Spread'].abs() < EXIT, 'Position'] = 0
full_pair_df.loc[full_pair_df['Spread'].abs() > STOP, 'Position'] = 0

# Forward-fill positions (hold trades)
full_pair_df['Position'] = full_pair_df['Position'].replace(0, np.nan).ffill().fillna(0)

```

```python
full_pair_df['Position_Lag'] = full_pair_df['Position'].shift(1)
full_pair_df['Position_Lag'] = (
    full_pair_df
    .groupby(['Stock1', 'Stock2'])['Position']
    .shift(1)
    .fillna(0)
)
full_pair_df['Pair_Return'] = (
    full_pair_df['Position_Lag'] *
    (
        full_pair_df['Weight_A'] * full_pair_df['Returns_Stock1']
        +
        full_pair_df['Weight_B'] * full_pair_df['Returns_Stock2']
    )
)
portfolio_returns = (
    full_pair_df
    .groupby('Date')['Pair_Return']
    .mean()  # equal weight per pair
)
equity_curve = (1 + portfolio_returns).cumprod()


sharpe = np.sqrt(252) * portfolio_returns.mean() / portfolio_returns.std()
max_dd = (equity_curve / equity_curve.cummax() - 1).min()

print("Sharpe:", sharpe)
print("Max Drawdown:", max_dd)
```

```python
TC = 0.001  # 10 bps

full_pair_df['Trade'] = full_pair_df['Position_Lag'].diff().abs()
full_pair_df['Cost'] = full_pair_df['Trade'] * TC

full_pair_df['Pair_Return_Net'] = full_pair_df['Pair_Return'] - full_pair_df['Cost']
# Compute net portfolio returns (after transaction costs)
portfolio_returns_net = (
    full_pair_df
    .groupby('Date')['Pair_Return_Net']
    .mean()
)

# Compute equity curve
equity_curve_net = (1 + portfolio_returns_net).cumprod()

# Plot gross vs net
equity_curve.plot(label='Gross')
equity_curve_net.plot(label='Net')
plt.legend()
plt.show()

```

```python
def perf_stats(returns):
    ann_ret = (1 + returns).prod()**(252/len(returns)) - 1
    ann_vol = np.sqrt(252) * returns.std()
    sharpe = ann_ret / ann_vol
    max_dd = (returns.cumprod() / returns.cumprod().cummax() - 1).min()
    return ann_ret, ann_vol, sharpe, max_dd

print("Gross:", perf_stats(portfolio_returns))
print("Net  :", perf_stats(portfolio_returns_net))
```

```python
pair_perf = (
    full_pair_df
    .groupby(['Stock1', 'Stock2'])['Pair_Return_Net']
    .sum()
    .sort_values()
)

print("Worst pairs:\n", pair_perf.head(10))
print("Best pairs:\n", pair_perf.tail(10))
```
